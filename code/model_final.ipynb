```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

def cgpa_calc(index):
    if index>8:
        return "Great Work"
    elif index>6:
        return "Good Work"
    elif index>4:
        return "Work Harder!"
    else: 
        return "Failed"
    

#Data extraction
mental_health_data = 'student_mental_health_burnout.csv'

data = pd.read_csv(mental_health_data)

processed_data = data.dropna()

processed_data.columns 

target = processed_data['cgpa']


features = ['age', 'daily_study_hours',
        'daily_sleep_hours', 'screen_time_hours','anxiety_score', 'depression_score', 'academic_pressure_score',
        'financial_stress_score', 'social_support_score',
        'physical_activity_hours', 'attendance_percentage',]

X = processed_data[features]
y = target

    #model creation - decision tree regressor
cgpa_model = DecisionTreeRegressor(random_state = 1)
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1)

    #training the model
cgpa_model.fit(train_X, train_y)

    #predicting values with the help of tree model
pred_val = cgpa_model.predict(test_X)
mae_tree = mean_absolute_error(pred_val, test_y)



```

```python
def Predict(info):
    input_data = [list(info)] 
    return cgpa_model.predict(input_data)[0]


data = input("Enter the age, Daily study time, Daily Sleep time, Screen time, anxiety score, depression_score, academic pressure score, \n financial stress score, social support score, physical activity hours, and finally attendance percentage, each with a space")


database = list(map(float, data.split()))


if len(database) != 11:
    print("Error: Please enter exactly 11 values separated by spaces.")
else:
    print(f"Predicted CGPA: {Predict(database):.2f}")

print(f"Mean Absolute Error obtained from DecisionTreeRegressor Model is {mae_tree}")
print(f"The relative accuracy of Decision Tree Regression model is {(100 - 100*(mae_tree/np.mean(test_y))):.2f}%")
```
